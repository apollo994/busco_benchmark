```python
species = [x.split('/')[-1].replace('.files','') for x in species]
```

```python
species_dict = {}
```

```python
# create species key and add referece genome

for sp in species:
    path = f'../data/{sp}.files/dna/{sp.capitalize()}.*.dna_rm.toplevel.fa*'
    files = ! ls $path | grep -v GCcontent
    
    if len(files)>1:
        species_dict[sp] = {"ref":files[0]}
    else:
        path = f'../data/{sp}.files/dna/{sp.capitalize()}.*.dna_rm.primary_assembly.fa*'
        print (sp)
        files = ! ls $path | grep -v GCcontent
        species_dict[sp] = {"ref":files[0]}
```

```python
# add annotation specific to each method 

methods = ['geneid', 'geneidblastx', 'glimmer', 'snap', 'augustus']

for method in methods:
    print (method)
    
    for sp in species_dict:
        path = ! ls /nfs/users/rg/fzanarello/projects/busco_benchmark/data/{sp}.files/{sp}.{method}.over200.gff3
        species_dict[sp][method] = path[0]
```

### To move from Ferriol folder to mine

```python
with open('move_and_gunzip.txt', 'w') as out:
    for sp in species_dict:

        
        ref_location_abs = os.path.abspath(species_dict[sp]['ref'])       
        file_name = os.path.basename(ref_location_abs)
        file_ext = os.path.splitext(file_name)[-1]
        
        target_dir = ' /nfs/users/rg/fzanarello/projects/busco_benchmark/genomes/'
        
        if file_ext == '.fa':
            out.write(f'cp {ref_location_abs} {target_dir}\n')
        
        if file_ext == '.gz':
            out.write(f'cp {ref_location_abs} {target_dir} && gunzip {target_dir}{file_name} \n')
            
        if file_ext != '.gz' and file_ext != '.fa':
            print (sp, species_dict[sp])
        
        
```

```python
annotaions_dict = {}

species = ! ls ../protein_sequences/

for sp in species:
    
    annotaions_dict[sp] = {}
    
    annotations = ! ls ../protein_sequences/$sp
    
    for ann in annotations:
        ann_prog = ann.split('.')[1]
        annotaions_dict[sp][ann_prog] = f'/users/rg/fzanarello/projects/busco_benchmark/protein_sequences/{sp}/{ann}'
        
```

```python
methods = ['geneid', 'geneidblastx', 'glimmer', 'snap', 'augustus']


with open ('../array_commands/busco_from_protein.txt', 'w') as out:
    for sp in annotaions_dict:
        
        for method in annotaions_dict[sp]:
        
            annotation = annotaions_dict[sp][method]
            ortho_db = 'vertebrata_odb10'
            out_folder = f'/raw_results/{sp}/{method}'
            
            out.write(f'/nfs/users/rg/fzanarello/projects/busco_benchmark/scripts/run_busco_from_protein.sh {annotation} {ortho_db} {out_folder} \n')
```
